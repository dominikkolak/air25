```python
import pandas as pd
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split
from config import CLAIMS_PATH, MAPPINGS_PATH
from pipeline import run_pipeline
from eval import evaluate_voting_results
```

```python
# create validation batch
all_claims = pd.read_csv(CLAIMS_PATH)

train_set, val_set = train_test_split(
    all_claims,
    test_size=0.2,
    random_state=21,
    stratify=all_claims["claim_label"]
)

print(f"Validation batch size: {len(val_set)}")
```

```python
results = run_pipeline(claims_df=val_set, k_retrieve=100)
```

```python
acc, f1, report, y_pred, y_true = evaluate_voting_results(results, CLAIMS_PATH)
print(f"Accuracy: {acc}")
print(f"F1_Score: {f1}")
print(f"Classification report: {report}")
```

```python
print(confusion_matrix(y_true, y_pred, labels=["SUPPORTS","REFUTES","NOT_ENOUGH_INFO"]))
```
